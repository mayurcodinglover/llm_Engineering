```python
# imports
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
openai=OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
```

```python
# here is the question; type over this to ask something new

system_prompt = """
You are a Sql expert and your work is to help users to write sql query User will give you a natural language and tell you what they wanted to do you have to answer it in sql query ex. i want to insert a row in a users table with name mayur and age 20 so you just need to give user insert query ok.
"""
```

```python
#get user prompt
def getUserPrompt(inputtxt):
    user_prompt=f"""
    give me sql query of below statement which is given and give me specific query and exact answer
    {inputtxt}
    """
    return user_prompt
```

```python
# Get Llama 3.2 to answer
def sqlQuery(inputtxt):
    response=openai.chat.completions.create(
        model=MODEL_LLAMA,
        messages=[
            {"role":"system","content":system_prompt},
            {"role":"user","content":getUserPrompt(inputtxt)}
        ],
    )
    result=response.choices[0].message.content
    return result;
```

```python
inputtxt=input("Enter your Questions");
print(sqlQuery(inputtxt))
```
